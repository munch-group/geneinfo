```python
%matplotlib inline
%config InlineBackend.figure_format='svg'

from geneinfo.plot import venn
from geneinfo.genelist import GeneList

import string
alphabet = GeneList(list(string.ascii_lowercase))
foo, bar, baz, obo, diz = (
    GeneList(list(alphabet[:14]) + ['aa', 'ae', 'oe']).name('foo'),
    GeneList(alphabet[1:10]).name('bar'),
    GeneList(alphabet[3:8]).name('baz'),
    GeneList(alphabet[13:15]).name('obo'),
    GeneList(alphabet[3:8]).name('diz'),
)
foo.fisher(bar, alphabet, return_counts=True)
```

```python
venn(foo, bar, background=alphabet, alpha=0.2, palette='rainbow')
```

```python
venn(foo, bar, baz, obo,
     petals=4, 
     background=alphabet,
     figsize=(7, 7),
     )
```

```python
venn(foo, bar, baz, obo, diz, 
     petals=5, 
     background=alphabet,
     figsize=(7, 7),
     )
```
