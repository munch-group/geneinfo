```python
from geneinfo.genelist import GeneList as glist
from geneinfo.plot import venn
```

## GeneList

Long lists of gene names do not work well visually:

```python
a = glist([f'gene{i}' for i in range(150)]).name('a')
b = glist([f'gene{i}' for i in range(0, 60)]).name('b')
c = glist([f'gene{i}' for i in range(30, 100)]).name('c')
d = glist([f'gene{i}' for i in range(0, 100, 2)]).name('d')
e = glist([f'gene{i}' for i in range(1, 100, 2)]).name('e')

venn(b, c, d, cmap='tab10', background=a)
```

```python
venn(b, c, d, e, cmap='tab10', background=a)
```

```python
venn(b, c, d, e, cmap='tab10', background=a, ncols=2)
```

```python
venn(b, c, d, e, cmap='tab10', background=a, petals=4)
```
