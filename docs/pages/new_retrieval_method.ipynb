```python
import geneinfo.utils as utils
gl = utils.GeneListCollection('google_sheet.csv')
names = gl.get('sfari_all_conf')
```

```python
from subprocess import Popen, PIPE
from collections import OrderedDict

fields = [
    'symbol',
    'gene-id',
    'ensembl-geneids',
    'name-id',
    'swissprot-accessions',
    'synonyms',
    'replaced-gene-id',
    'gene-type',
    'chromosomes',
    'genomic-region-gene-range-range-start',
    'genomic-region-gene-range-range-stop',
    'summary-description',
    'annotation-assembly-name',
]

assembly_names = {'hg38': 'GRCh38',
                  'GRCh38': 'GRCh38',
                  'T2T': 'T2T',
                  't2t': 'T2T'
                 }
assembly = 'hg38'

cmd = f"datasets summary gene symbol {' '.join(names)} --as-json-lines"
p = Popen(cmd.split(), stdout=PIPE)
cmd = f'dataformat tsv gene --elide-header --fields {",".join(fields)}'
output = subprocess.check_output(cmd.split(), stdin=p.stdout)
# print(output)
p.wait()
for line in output.decode().strip().split('\n'):

    d = dict(zip(fields, line.strip().split('\t')))

    if 'annotation-assembly-name' in d and d['annotation-assembly-name'].startswith(assembly_names[assembly]):

        info = {}
        info['name'] = d['symbol']
        info['ensembl'] = d['ensembl-geneids']
        info['ncbi'] = d['name-id']
        info['swissprot'] = d['swissprot-accessions']
        info['synonyms'] = d['synonyms'].split(',')
        info['_ncbi'] = d['replaced-gene-id']
        info['type'] = d['gene-type']
        info['assembly']: d['annotation-assembly-name']
        info['coord'] = (d['chromosomes'], d['genomic-region-gene-range-range-start'], d['genomic-region-gene-range-range-stop'])
        
    print(info)
```
