
**geneinfo** is a compilation of gene information APIs adapted for use in a Jupyter notebook and some additional utilities. Its [API](./api/) lets you retrieve of information about genes and gene ontologies, plot data together with gene structures (genome browser style), show data on chromosome ideograms, perform gene ontology enrichment analyses and more.

## Interpreting data in a gene context

Among many other things, `geneinfo` lets you plot data in chromosome coordinates and view overlapping gene annotation using [](`~geneinfo.plot.gene_plot`).

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

# %config InlineBackend.figure_format = 'retina'
%config InlineBackend.figure_format = 'svg'
````

Dummy data:

````python
x = np.linspace(76500000, 77500000, 10000)
rng = np.random.default_rng()
y = rng.exponential(scale=1, size=x.size)
````

````python
import geneinfo.plot as gplt
ax = gplt.gene_plot('chrX', 76500000, 77500000, assembly='hg19')
ax.scatter(x, y, s=2, c='tab:blue') ;
````

You can get brief summaries using [](`~geneinfo.information.gene_info`).

````python
import geneinfo.information as gi
````

````python
gi.gene_info(['ATRX', 'MAGT1'])
````

Get gene ontology terms associated with genes.

````python
import geneinfo.ontology as go
go.email("youremail@address.com")
````

````python
go_terms = go.get_go_terms_for_genes(['ATRX', 'MAGT1'], evidence=['IDA', 'EXP', 'HTP'])
go_terms[:10]
````

Show graph for those terms.

````python
go.show_go_dag_for_terms(go_terms)
````

Show definitions of those terms.

````python
go.go_info(go_terms)
````
